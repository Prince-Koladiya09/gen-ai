```python
import mediapy as media
import random
import torch
from diffusers import DiffusionPipeline
```

```python
pipe=DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0",torch_dtype=torch.float16)
pipe.to("cuda")
```

```python
from PIL import Image

prompt='space'
negative_prompt=negative_prompt = "bad-picture-chill-75v, ng_deepnegative_v1_75t, badhandv4, (worst quality:2), (low quality:2), (normal quality:2), (lowres:2), (bad anatomy:2), (bad hands:2), (watermark:2), (mole:1.5), (freckles:1.5)"
width=int(512)
heigtht=int(512)
seed=random.randint(0,2147483647)

image=pipe(prompt=prompt,
           guidance_scal=20.0,
           width=width,
           hight=heigtht,
           negative_prompt=negative_prompt,
           generator=torch.Generator().manual_seed(seed),
           num_images_per_prompt=1,
           num_inference_steps=50,
           output_type="tensor",
           ).images

pil_image = Image.fromarray(image[0].astype('uint8'))
pil_image.save("output.jpg")

media.show_images(image)
```
